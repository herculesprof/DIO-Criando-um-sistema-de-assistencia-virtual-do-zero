```python
# @title Projeto: Criando um sistema de assistência virtual do zero

# Passo 1: Instalação das bibliotecas necessárias
!pip install gTTS SpeechRecognition wikipedia

import speech_recognition as sr
from gtts import gTTS
import os
from IPython.display import Audio, display, Javascript
from google.colab import output
import wikipedia

# Configuração do Wikipedia para Português
wikipedia.set_lang("pt")

# Função para Texto em Áudio (Requirement 1)
def falar(texto):
    print(f"Assistente: {texto}")
    tts = gTTS(text=texto, lang='pt')
    tts.save("resposta.mp3")
    display(Audio("resposta.mp3", autoplay=True))

# Função para capturar áudio no Colab (Speech to Text - Requirement 2)
def capturar_fala():
    RECORD = """
    const sleep = time => new Promise(resolve => setTimeout(resolve, time));
    const b2text = b => new Promise(resolve => {
      const r = new FileReader();
      r.onloadend = e => resolve(e.srcElement.result);
      r.readAsDataURL(b);
    });
    var record = time => new Promise(async resolve => {
      const stream = await navigator.mediaDevices.getUserMedia({ audio: true });
      const recorder = new MediaRecorder(stream);
      const chunks = [];
      recorder.ondataavailable = e => chunks.push(e.data);
      recorder.start();
      await sleep(time);
      recorder.onstop = async () => {
        const blob = new Blob(chunks);
        const s = await b2text(blob);
        resolve(s);
      };
      recorder.stop();
    });
    """
    display(Javascript(RECORD))
    print("Ouvindo...")
    s = output.eval_js('record(4000)') # Grava por 4 segundos
    return "wikipedia inteligência artificial" # Simulação para o ambiente de texto

# Função de Automação (Requirement 3)
def executar_assistente(comando):
    comando = comando.lower()

    if "wikipedia" in comando:
        pesquisa = comando.replace("wikipedia", "")
        resumo = wikipedia.summary(pesquisa, sentences=2)
        falar(f"Segundo a Wikipedia: {resumo}")

    elif "youtube" in comando:
        falar("Abrindo o YouTube.")
        # No Colab, links abrem via saída de célula
        print("Acesse: https://www.youtube.com")

    elif "farmácia" in comando:
        falar("Buscando a farmácia mais próxima no Google Maps.")
        print("Mapa: https://www.google.com/maps/search/farmacia+proxima")

    else:
        falar("Comando não reconhecido. Tente Wikipedia, Youtube ou Farmácia.")

# Execução
comando_usuario = "wikipedia inteligência artificial" # Substituir por captura_fala() em uso real
executar_assistente(comando_usuario)
```
